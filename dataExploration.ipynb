```python
import pandas as pd
import numpy as np
import pickle
```

In this file we will explore the already cleaned data, which we did in "dataCleaning.ipynb"

```python
# Loading the data
with open('Data/Dataframes/newDF.pkl', 'rb') as f:
    df = pickle.load(f)
    
    
# Dropping ID and onehot columns for visualization
newDF = df[['Resume', 'Label', 'Skills', 'TextLen']]

print(newDF.columns)
```

We will quickly look at the spread of the data in the columns. As we are working with resumes as text, we cant get many interesting visualizations yet.

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
newDF['Skills'].value_counts()
```

```python
# Lets see the distribution of the labels
plt.figure(figsize=(10, 5))
sns.countplot(x='Skills', data=newDF, palette='Set2')
plt.title('Distribution of Labels')
plt.xlabel('Label')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
